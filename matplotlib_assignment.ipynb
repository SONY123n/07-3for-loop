```python
#matplotlib is a library used for analysis the data
#1)-scatter plot: scatterplot is used to understand thee relationship between two variables
#2)- barplot : it is used to represent the categorical data 
#3)-pieplot : it is used for the distribution of the different unique classes
#4-histogram/histplot : is used to the distribution of the continuous data
#5)-pairplot :it is used to represent the all the numerical data in one plot
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#scatter plot: scatterplot is used to understand thee relationship between two variables

np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

```python
#subplot:it allows to plot multiple plot 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.plot(x,y)
plt.show()
```

```python
# barplot : it is used to represent the categorical data 

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

x = np.arange(len(company))
plt.bar(x, profit)
plt.xticks(x, company)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Profits')
plt.show()
```

```python
#boxplot:is used to detremoine the outlier

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], notch=True)  

plt.xlabel('Box Labels')
plt.ylabel('Values')
plt.title('Comparison of Box Distributions')
plt.grid(True) 

plt.show()
```
