```python
tit=sns.load_dataset('titanic')
tit
```

```python
sns.boxplot(x='pclass', y='age', data=tit, palette='pastel', notch=True)
plt.title('Age Distribution by Class with Notches')
plt.show()


sns.boxplot(x='class', y='fare', data=tit, palette='pastel', notch=True)
plt.title('Age Distribution by Class with Notches')
plt.show()
```

```python
dia=sns.load_dataset('diamonds')
dia
```

```python
sns.histplot(x='price',hue='cut',data=dia)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Prices by Cut')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

flights = sns.load_dataset("flights")

# Convert 'month' column to numerical representation
flights['month'] = pd.Categorical(flights['month']).codes + 1  # +1 to avoid 0
flights_pivot = flights.pivot(index="month", columns="year", values="passengers") 

sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="Oranges")
plt.title("Monthly Passenger Counts by Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
